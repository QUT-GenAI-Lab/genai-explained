# Introduction to Google Colab

If you've never used Google Colab before to run interactive code online - then start here!

These notebooks have been designed to be used with the free tier of the Google Colab cloud platform, specifically the 'T4' instance (essentially a free GPU). There are also python package resources built into some of these notebooks that need to be installed, but they're a little funny to install on Google Colab. Here's some general housekeeping.

## Some general notes about python notebooks:
If you're new to python and/or python notebooks - firstly, welcome! It's a very accessible coding language, and if you're interested in learning how to code, it's a great option to start with.

Unique to python is a format called "python notebooks", which is what these tutorials are written in. Essentially, python notebooks allow us to run chunks of code, called "cells", one at a time, and then look at any outputs from those cells immediately:

![Image of a code cell and the generated output](./assets/0000-celloutputdiag.png)

To run a code cell in Google Colab, simply hover over the cell and then click the little play button that pops up. Either that, or you can hit Shift+Enter on your keyboard:

![Image of a code cell highlighting the 'play' button](./assets/0000-celldiag.png)

**A BIG NOTE:**
Python (like almost all programming languages) is designed to execute its logic sequentially from top to bottom. Because of this, with these python notebooks, **you must run ALL cells prior to the one you are interested in running.** An easy way to do this by habit is simply hitting Shift+Enter at each cell from the top of the notebook. This automatically scrolls the notebook, and also runs each cell as you go along. If you're having trouble with code not running correctly in your notebook, the first thing you should try is to run everything from top to bottom with no skips. You can do this easily by selecting the "Run all" option from the Runtime dropdown menu:

![Image of a drop-down box highlighting the 'Run all' option](./assets/0000-runalldemo.png)

## GPU Instance:
To make sure the code in these notebooks runs properly, make sure that when you're on Google Colab, you are on a T4 instance. To confirm that you're on one, just check the top right corner of your browser window when you're in google colab. Once you've run a cell, or have clicked "connect", it'll look like this:

![Image showing how to check the 'T4' runtime instance is being used](./assets/0000-t4check.png)

**Specifically, look for the "T4" in the top right corner.**

If not, you'll need to switch to the correct runtime. To do so, you'll need to click the drop down arrow in the right hand corner, and click "change runtime type":

![Image showing how to find the 'change runtime' menu in Google Colab](./assets/0000-runtimetypechange.png)

From there, you'll get a window pop-up that lets you pick which runtime to use. Under "Hardware Accelerator", pick, "T4 GPU", and click Save:

![Image showing how to select the T4 runtime in Google Colab](./assets/0000-t4select.png)

And then you should be on a T4 instance!

**A NOTE:** on the free tier of Google Colab, you will get a limited amount of compute time on a T4 instance every month (which is nebulously defined by Google dynamically, based on server load). If a T4 instance is unavailable, theoretically these notebooks should still work, but much much slower.

If you are comfortable with python and Jupyter Notebooks (the format of these tutorials), feel free to download these notebooks and run them locally! 

## Package installs:
For the sake of keeping repeated code to a minimum, a lot of functions, variables, and pieces of code that are used across multiple notebooks are installed in a package called SDutils. For the notebooks that need it, there'll be a cell at the top which installs it, that looks like this:

![Image showing a 'pip' install command](./assets/0000-pipinstall.png)


However, because this needs to be installed into the runtime, Google Colab in particular seems to struggle with acknowledging its installation the first time around. In particular, you'll get this error:

![Image showing a 'Module not found' error](./assets/0000-nomodule.png)

If this happens (which it often does), simply go to the top menu ribbon, and under "Runtime", click either "Restart Session" (if you want to run each cell one by one by yourself), or "Restart Session and Run All" (if you just want to see all outputs):

![Image showing how to restart the current Google Colab session](./assets/0000-restartsession.png)

There will be a dialogue box that pops up asking you to confirm your selection, but after that, the notebook should work!
